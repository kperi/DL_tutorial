```python
import numpy  as np
from scipy.ndimage import convolve
from PIL import Image
%matplotlib inline
import matplotlib.pyplot as plt


img = np.asarray( Image.open('media/flower.jpg'))
gimg = np.asarray( Image.open('media/flower.jpg').convert('L') )


k0 = np.array([[1, 0, -1],
              [0, 0, 0],
              [-1, 0, 1]])

k0 = np.array([[0, 1, 0],
              [1, -4, 1],
              [0, 1, 0]])

k2 = np.array([[1, 1, 1],
              [1, 1, 1],
              [1, 1, 1]], dtype=np.float)/9


sobel = np.array([[-1, -2, -1],
              [0, 0, 0],
              [1, 2, 1]], dtype=np.float)

 
sobel.T
```

## Original image

```python
plt.imshow(img)
```

## Grayscale image

```python
plt.imshow(gimg, cmap=plt.get_cmap('gray'))
```

## Edge detection

```python
img1 = convolve(gimg, k1, mode='constant')
plt.imshow(img1, cmap=plt.get_cmap('gray'))
```

```python
img2 = convolve(gimg, k0, mode='constant')
plt.imshow(img2, cmap=plt.get_cmap('gray'))
```

## Box blur

```python
img2 = convolve(gimg, k2, mode='constant')
plt.imshow(img2, cmap=plt.get_cmap('gray'))
```

## SobelX

```python
sobelx = convolve(gimg, sobel, mode='constant')
plt.imshow(sobelx, cmap=plt.get_cmap('gray'))
```

## SobelY

```python
sobely = convolve(gimg, sobel.T, mode='constant')
plt.imshow(sobely, cmap=plt.get_cmap('gray'))
```
